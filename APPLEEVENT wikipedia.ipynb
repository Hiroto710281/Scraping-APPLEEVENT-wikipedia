```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
url ='https://ja.wikipedia.org/wiki/Apple_Special_Event'
```

```python
result = requests.get(url)
bs = BeautifulSoup(result.text, 'html')
for a in bs.find_all('a'):
    if a.string != None:
        print(a.string.strip())
        href = a.get('href')
        if href != None:
            if href != None:
                if href.startswith('http'):
                    print(href)
                else:
                    print('https://ja.wikipedia.org/wiki/Python' + href)
                print()
```

```python
result = requests.get(url)
bs = BeautifulSoup(result.text, 'html')
for h2 in bs.find_all('h2'):
    if h2.string != None:
        print(h2.string)
        print()
```
